```python
import pandas as pd

# df = pd.read_csv('tc_text_mining_mpcif_curie.csv')
df = pd.read_csv('tc_text_mining_mpcif.csv')
df = df.drop('Unnamed: 0',axis=1)
df['dum'] = range(len(df))
```

```python
import os 
from os import listdir
from shutil import copyfile

dirs = './temperature_data_rare_earth_free_curie_mag_graph'
# dirs = './test'

# short_num = 10

name = dirs + '/' + 'id_prop.csv'
os.makedirs(dirs,exist_ok=True)
# type_df = df.loc[:,['dum','temperature']].iloc[:short_num]
type_df = df.loc[:,['dum','temperature']]
type_df.columns = ['ID','target']
type_df = type_df.reset_index()
type_df = type_df.loc[:,['ID','target']]
type_df.to_csv(name,index=False,header=False)

type_id_df = df.loc[:,['name','dum']]
cifs_path = './CIFs/'
mag_cifs = []
for form_id, form in enumerate(type_id_df.loc[:,'name'].to_list()):
    cif_form_path = cifs_path + form    
#     listdir(cif_form_path)
    directory = [i for i in listdir(cif_form_path) if i[-7:] == 'mag.cif']
#     directory = [i for i in directory if i[-3:] == 'cif']
    if len(directory) != 0: 
        mag_cifs.append(form_id)
        mag_cif = directory[0]
        src = cif_form_path + '/' + mag_cif
        dst = dirs + '/' + str(form_id) + '.cif'
        copyfile(src, dst)
    else:
        continue
type_df = type_df.loc[mag_cifs, :]
type_df.to_csv(name,index=False,header=False)
```
