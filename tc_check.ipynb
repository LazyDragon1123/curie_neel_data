```python
def name_converter(case):
    to_list = case[2:-2]
    splits = to_list.split('","')
    return splits, len(splits)
```

```python
def filtering(names_list,names_len,reverse=False):
    res_list = []
    res_len = []
    cout = 0
    for names, len_name in zip(names_list, names_len):
        if reverse:
            names.reverse()
        if len_name > 1:
            name_this = [names[0]]
            for cand in names[1:]:
                if ' ' in cand:
                    continue
                elif ',' in cand:
                    continue
                elif '-' in cand:
                    continue
                elif '[' in cand:
                    continue
                elif '"\""' in cand:
                    continue
                elif '/' in cand:
                    continue
                elif '+' in cand:
                    continue
                elif 'δ' in cand:
                    continue
                elif '−' in cand:
                    continue
                elif '(' in cand:
                    continue
                elif '2019' in cand:
                    continue
                elif 'MX' in cand:
                    continue
                elif 'μ' in cand:
                    continue
                elif 'μ' in cand:
                    continue
                elif 'μ' in cand:
                    continue
                elif cand[1:3].islower():      
                    continue
                elif len(cand) > 1 and cand.isupper():
                    continue
                else:
                    name_this.append(cand)

            res_list.append(name_this)

            res_len.append(len(name_this))
            
        else:
            res_list.append(names)
            res_len.append(1)
    return res_list, res_len
```

```python
def base_filter(df):
    formula = df.index.to_list()
    indices = []
    for ind, form in enumerate(formula):
        if type(form) != str:
            print(form)
            continue
        else:
            indices.append(ind)           
    res_df = df.copy().iloc[indices,:]
    return res_df, indices
def O_filter(df,ele='O'):
    formula = df.index.to_list()
    indices = []
    for ind, form in enumerate(formula):
        if type(form) != str:
            print(form)
            continue
        elif ele in form:
            indices.append(ind)
        else:
            continue            
    res_df = df.copy().iloc[indices,:]
    return res_df, indices

def zero_filter(df):
    formula = df.index.to_list()
    indices = []
    for ind, form in enumerate(formula):
        if type(form) != str:
            print(form)
            continue
        elif '0' not in form:
            indices.append(ind)
        else:
            continue            
    res_df = df.copy().iloc[indices,:]
    return res_df, indices

def symbol__filter(df):
    formula = df.index.to_list()
    indices = []
    for ind, form in enumerate(formula):
        if type(form) != str:
            print(form)
            continue
        elif form == '':
            continue
        elif (' ' not in form) and ('+' not in form) and ('[' not in form) and ("'" not in form) and ('·' not in form) and ('−' not in form) and ('–' not in form) and (')' not in form) and ('/' not in form) and ('.' not in form) and ('-' not in form)  and (not form.isupper()) and (not form.islower()):
            indices.append(ind)
        else:
            continue            
    res_df = df.copy().iloc[indices,:]
    return res_df, indices

def exclude_filter(df,ele='Sm'):
    formula = df.index.to_list()
    indices = []
    for ind, form in enumerate(formula):
        if type(form) != str:
            print(form)
            continue
        elif ele not in form:
            indices.append(ind)
        else:
            continue            
    res_df = df.copy().iloc[indices,:]
    return res_df, indices
```

```python
oxides_formulas = list(set(mod_data.index.to_list()))
df_dict = {}
for name in oxides_formulas:
    df_dict[name] = pd.DataFrame()
    
import re
import numpy as np
from collections import Counter

def str_num(s):
    if type(s) == int:
        return s
    else:
        num_list = re.findall(r"\d+", s)
        num_list = np.array([int(i) for i in num_list])
#         ave = np.mean(num_list)
        fir = num_list[0]
        return fir
```

```python
# concat
for ind, form in enumerate(oxides_formulas):
    if ind/1000 == ind//1000:
        print(ind)
    each_df = mod_data.copy().loc[form,:]
    Type = 'Curie'
    temp = 1
    flag = False
    if mod_data.index.to_list().count(form) == 1:
        Type = each_df.loc['Type']
        temp = each_df.loc['Normalised Value']
        try:
            temp = str_num(temp)
        except:
            flag = True
    else:
        Types = each_df.loc[:,'Type'].to_list()
        counter = Counter(Types)
        Type = counter.most_common()[0][0]
        temps = each_df.loc[:,'Normalised Value']
        temps = np.array([str_num(i) for i in temps])
        temp = np.mean(temps)

    temp_df = pd.DataFrame(index=[ind])
    temp_df['Type'] = Type
    temp_df['Normalised Value'] = temp
    temp_df['flag'] = flag
    
    df_dict[form] = temp_df
```

```python
saved_data = pd.concat(df_dict,axis=0)
```
