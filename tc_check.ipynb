```python
def name_converter(case):
    to_list = case[2:-2]
    splits = to_list.split('","')
    return splits, len(splits)
```

```python
names_list = [name_converter(i)[0] for i in names] 
names_len = [name_converter(i)[1] for i in names] 
```

```python
def filtering(names_list,names_len,reverse=False):
    res_list = []
    res_len = []
    cout = 0
    for names, len_name in zip(names_list, names_len):
        if reverse:
            names.reverse()
        if len_name > 1:
            name_this = [names[0]]
            for cand in names[1:]:
                if ' ' in cand:
                    continue
                elif ',' in cand:
                    continue
                elif '-' in cand:
                    continue
                elif '[' in cand:
                    continue
                elif '"\""' in cand:
                    continue
                elif '/' in cand:
                    continue
                elif '+' in cand:
                    continue
                elif 'δ' in cand:
                    continue
                elif '−' in cand:
                    continue
                elif '(' in cand:
                    continue
                elif '2019' in cand:
                    continue
                elif 'MX' in cand:
                    continue
                elif 'μ' in cand:
                    continue
                elif 'μ' in cand:
                    continue
                elif 'μ' in cand:
                    continue
                elif cand[1:3].islower():      
                    continue
                elif len(cand) > 1 and cand.isupper():
                    continue
                else:
                    name_this.append(cand)

            res_list.append(name_this)

            res_len.append(len(name_this))
            
        else:
            res_list.append(names)
            res_len.append(1)
    return res_list, res_len
```

```python
def O_filter(df):
    formula = df.index.to_list()
    indices = []
    for ind, form in enumerate(formula):
        if type(form) != str:
            print(form)
            continue
        elif 'O' in form:
            indices.append(ind)
        else:
            continue            
    res_df = df.copy().iloc[indices,:]
    return res_df, indices

def zero_filter(df):
    formula = df.index.to_list()
    indices = []
    for ind, form in enumerate(formula):
        if type(form) != str:
            print(form)
            continue
        elif '0' not in form:
            indices.append(ind)
        else:
            continue            
    res_df = df.copy().iloc[indices,:]
    return res_df, indices

def symbol__filter(df):
    formula = df.index.to_list()
    indices = []
    for ind, form in enumerate(formula):
        if type(form) != str:
            print(form)
            continue
        elif (' ' not in form) and ('+' not in form) and ('−' not in form) and ('–' not in form) and (')' not in form) and ('/' not in form) and ('.' not in form) and ('-' not in form)  and (not form.isupper()):
            indices.append(ind)
        else:
            continue            
    res_df = df.copy().iloc[indices,:]
    return res_df, indices
```

```python
mod_data, _ = O_filter(mod_data)
print(len(mod_data))
mod_data, _ = zero_filter(mod_data)
print(len(mod_data))
mod_data, _ = symbol__filter(mod_data)
print(len(mod_data))
```
