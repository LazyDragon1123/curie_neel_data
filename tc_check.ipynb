```python
import pandas as pd
path = './csv/phase_transitions.csv'
data = pd.read_csv(path)
```

```python
conf = data.loc[:,'Confidence '].to_list()
type_mag = data.loc[:,'Type'].to_list()
names = data.loc[:,'Names'].to_list()
```

```python
import numpy as np
cout = 0
for i in conf:
    if np.isnan(i):
        cout += 1
        
print(cout)
print('out of ')
print(len(conf))
```

```python
cout_c = 0
cout_n = 0
cout_o = 0
for i in type_mag:
    if i == 'Curie':
        cout_c += 1
    elif i == 'Néel':
        cout_n += 1
    else:
        cout_o += 1
        
print('curie')
print(cout_c)
print('neel')
print(cout_n)
print('other')
print(cout_o)
print('out of ')
print(len(type_mag))
```

```python
def name_converter(case):
    to_list = case[2:-2]
    splits = to_list.split('","')
    return splits, len(splits)
```

```python
names_list = [name_converter(i)[0] for i in names] 
names_len = [name_converter(i)[1] for i in names] 
```

```python
cout = 0 
for i in names_len:
    if i == 1:
        cout += 1
cout_f = len(names_list) - cout 
print(cout)
print(cout_f)
```
